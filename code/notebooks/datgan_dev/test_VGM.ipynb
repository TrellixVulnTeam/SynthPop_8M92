# Preprocess with DATGAN

```python
continuous_columns = ["distance", "age", "departure_time"]
```

```python
# personalised graph
graph = nx.DiGraph()
graph.add_edges_from([("age", "license"),
                      ("age", "education_level"),
                      ("gender", "work_status"),
                      ("education_level", "work_status"),
                      ("work_status", "hh_income"),
                      ("education_level", "hh_income"),
                      ("hh_income", "hh_descr"),
                      ("hh_income", "hh_size"),
                      ("hh_size", "hh_vehicles"),
                      ("hh_size", "hh_bikes"),
                      ("work_status", "trip_purpose"),
                      ("trip_purpose", "departure_time"),
                      ("trip_purpose", "distance"),
                      ("travel_dow", "choice"),
                      ("distance", "choice"),
                      ("departure_time", "choice"),
                      ("hh_vehicles", "choice"),
                      ("hh_bikes", "choice"),
                      ("license", "choice"),
                      # Non necessary links
                      ("education_level", "hh_size"),
                      ("work_status", "hh_descr"),
                      ("work_status", "hh_size"),
                      ("hh_income", "hh_bikes"),
                      ("hh_income", "hh_vehicles"),
                      ("trip_purpose", "choice")])

#graph = nx.DiGraph()
#graph.add_nodes_from(['age'])

output_folder = '../output/DATGAN_TEST/'
```

```python
datgan = DATGAN(continuous_columns, max_epoch = 2, steps_per_epoch=2000, batch_size=200, output=output_folder, gpu=0)
```

```python
datgan.fit(df, graph)
```

```python
test = datgan.sample(200)
```

```python
test
```

```python
datgan.dag = graph
datgan.verify_dag(df)
var_order = datgan.get_order_variables()
preprocessor = Preprocessor(continuous_columns=continuous_columns, 
                                 columns_order=var_order)
data = preprocessor.fit_transform(df)
```

# Preprocess with CTGAN

```python
ctgan = CTGANSynthesizer(verbose=True, cuda=True, batch_size=200)
```

```python
from ctgan.data_transformer import DataTransformer
```

```python
ctgan._validate_discrete_columns(df, discrete_columns)
ctgan._transformer = DataTransformer()
ctgan._transformer.fit(df, discrete_columns)
train_data = ctgan._transformer.transform(df)
```
