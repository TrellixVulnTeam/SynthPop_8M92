```python
input_folder = '../output/'
output_folder = '../synth_data/'
```

# Generate data from TGAN

```python
from tgan.model import TGANModel

tgans = ['TGAN_100', 'TGAN_200', 'TGAN_500', 'TGAN_1000']
```

```python
for name in tgans:
    tgan = TGANModel.load(input_folder + name + '/trained')
    samples = tgan.sample(len(df))
    samples.age = np.round(samples.age)
    samples.to_csv(output_folder + name + '.csv', index=False)
```
