```python
import os
os.chdir('../..')
```

```python
import numpy as np
import pandas as pd

import random

from tgan.model import TGANModel

import tensorflow as tf

import json

# For the Python notebook
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

# Load the data

```python
df = pd.read_csv('../data/Chicago/full.csv')
```

```python
continuous_columns = ['distance', 'age', 'departure_time']
```

```python
cols = list(df.columns)
```

# Specify some parameters

```python
output_folder = './output_TGAN/columns_order/'
max_epoch = 100
steps_per_epoch = 2000
```

```python
# Load the orders.json

with open(output_folder + 'orders.json', 'r') as infile:
    orders = json.load(infile)
```

# Prepare JSON

```python
orders = {}
```

# Train TGAN with original order

```python
idx_cont = [cols.index(cc) for cc in continuous_columns]
idx_cont.sort()
```

```python
orders['original'] = cols

with open(output_folder + 'orders.json', 'w') as outfile:
    json.dump(orders, outfile)
```

```python
output = output_folder + 'original/'
try:
    os.mkdir(output)
except:
    pass
```

```python
tgan = TGANModel(idx_cont, max_epoch=max_epoch, 
                 steps_per_epoch=steps_per_epoch,
                 output=output)

tgan.fit(df)
tgan.save(output + 'trained', force=True)

del tgan
df.columns = cols
```

# Train TGAN with "best order"

```python
best_order = ['age', 'gender', 'education_level', 'work_status', 
              'license', 'travel_dow', 'hh_descr', 'hh_income', 
              'hh_size', 'hh_vehicles', 'hh_bikes', 'trip_purpose',
              'distance', 'departure_time', 'choice']

# Put travel_dow first

df = df[best_order]
cols = best_order
```

```python
idx_cont = [best_order.index(cc) for cc in continuous_columns]
idx_cont.sort()
```

```python
orders['best'] = cols

with open(output_folder + 'orders.json', 'w') as outfile:
    json.dump(orders, outfile)
```

```python
output = output_folder + 'best/'
try:
    os.mkdir(output)
except:
    pass
```

```python
tf.reset_default_graph()

tgan = TGANModel(idx_cont, max_epoch=max_epoch, 
                 steps_per_epoch=steps_per_epoch,
                 output=output)

tgan.fit(df)
tgan.save(output + 'trained', force=True)

del tgan
df.columns = cols
```

# Train TGAN with random order

```python
for i in range(1, 10):

    cols = list(df.columns)
    random.shuffle(cols, random.random)
    
    df = df[cols]
    
    idx_cont = [cols.index(cc) for cc in continuous_columns]
    idx_cont.sort()
    
    name = 'order' + str(i+1)
    
    orders[name] = cols

    with open(output_folder + 'orders.json', 'w') as outfile:
        json.dump(orders, outfile)
        
    output = output_folder + name + '/'
    try:
        os.mkdir(output)
    except:
        pass
        
    tf.reset_default_graph()

    tgan = TGANModel(idx_cont, max_epoch=max_epoch, 
                     steps_per_epoch=steps_per_epoch,
                     output=output)
    
    tgan.fit(df)
    tgan.save(output + 'trained', force=True)

    del tgan
    df.columns = cols
```
