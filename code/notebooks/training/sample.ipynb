```python
dataset = 'Chicago'
name = 'NOLINKS'
```

```python
df = pd.read_csv('../data/' + dataset + '/data.csv')
```

```python
input_folder = '../output/' + dataset + '/'
output_folder = '../synth_data/' + dataset + '/'
```

```python
if name not in os.listdir(input_folder):
    raise ValueError("Name of the model unknown!")
```

```python
model = None
if name is 'CTGAN':
    import ctgan
    from ctgan import CTGANSynthesizer
    
    ctgan = CTGANSynthesizer()
    model = ctgan.load(input_folder + 'CTGAN/trained.pkl')
    
elif name is 'TGAN':
    from tgan.model import TGANModel
    model = TGANModel.load(input_folder + name + '/trained')
else:
    
    LIB = None
        
    if 'WGAN' in name:
        from modules.datgan import DATWGAN as LIB
    if any(x in name for x in ('WGANGP','WGGP')) :
        from modules.datgan import DATWGANGP as LIB
    elif 'DRAGAN' in name:
        from modules.datgan import DATDRAGAN as LIB
    else:
        from modules.datgan import DATSGAN as LIB


    model = LIB.load(input_folder + name + '/', 'trained')
```

```python
samples = model.sample(len(df))
if dataset is 'Chicago':
    samples.age = np.round(samples.age)
elif dataset is 'LPMC':
    samples.age = np.round(samples.age)
    samples.distance = np.round(samples.distance)
samples.to_csv(output_folder + name + '.csv', index=False)
```
