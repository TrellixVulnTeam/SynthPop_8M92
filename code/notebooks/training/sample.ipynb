```python
input_folder = '../output/' + dataset + '/'
output_folder = '../synth_data/' + dataset + '/'
```

```python
if name not in os.listdir(input_folder):
    raise ValueError("Name of the model unknown!")
```

```python
model = None
if name is 'CTGAN':
    import ctgan
    from ctgan import CTGANSynthesizer
    
    ctgan = CTGANSynthesizer()
    model = ctgan.load(input_folder + 'CTGAN/trained.pkl')
    
elif name is 'TGAN':
    from tgan.model import TGANModel
    model = TGANModel.load(input_folder + name + '/trained')
else:
    
    LIB = None
        
    if 'PRELSTM' in name:
        from modules.datgan_preLSTM.model import DATGAN as LIB
    elif 'POSTLSTM' in name:
        from modules.datgan_postLSTM.model import DATGAN as LIB
    else:
        from modules.datgan.model import DATGAN as LIB


    model = LIB.load(input_folder + name + '/', 'trained')
```

```python
samples = model.sample(len(df))
if dataset is 'Chicago':
    samples.age = np.round(samples.age)
elif dataset is 'LPMC':
    samples.age = np.round(samples.age)
    samples.distance = np.round(samples.distance)
samples.to_csv(output_folder + name + '.csv', index=False)
```

# TEST

```python
#output_folder = '../output/' + dataset + '/' + name + '/'
output_folder = '../output/' + dataset + '/DATGAN_1k3/' # Model with argmax

with open(output_folder + 'logs/stats.json', 'r') as infile:
    data = json.load(infile)
    
loss = {'gen': [],
        'discr': []
       }

for d in data:
    loss['gen'].append(d['GAN_loss/gen/final-g-loss'])
    loss['discr'].append(d['GAN_loss/discrim/loss'])
    
plt.figure(figsize=(10,7))
plt.plot(loss['gen'], 'k', label='Generator')
plt.plot(loss['discr'], 'r', label='Discriminator')

plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
```
