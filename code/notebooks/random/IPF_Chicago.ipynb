```python
import numpy as np
import pandas as pd

from modules.ipf import ipf

import matplotlib.pyplot as plt

# For the Python notebook
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv('../data/Chicago/full_num.csv')
```

```python
df.head()
```

```python
cols = np.array(df.sum(axis=0))
rows = np.array(df.sum(axis=1))
```

```python
ipf = ipf()
```

```python
mat = ipf.fit(initial_joint = np.ones((len(rows), len(cols))), marginal1 = rows, marginal2 = cols, maxitr = 10, method = 'mle')
```

```python
synth_df1 = pd.DataFrame(mat)
synth_df1.columns = df.columns
```

```python
synth_df1.head()
```

```python
mat = ipf.fit(initial_joint = df, marginal1 = rows, marginal2 = cols, maxitr = 10, method = 'mle')
```

```python
synth_df2 = pd.DataFrame(mat)
synth_df2.columns = df.columns
```

```python
synth_df2.head()
```

```python
mu, sigma = 0, 0.1 
# creating a noise with the same dimension as the dataset
noise = np.random.normal(mu, sigma, [len(df), len(df.columns)]) 
print(noise)
```

```python
noisy_df = df + noise
```

```python
noisy_df.head()
```

```python
mat = ipf.fit(initial_joint = noisy_df, marginal1 = rows, marginal2 = cols, maxitr = 10, method = 'mle')
```

```python
synth_df3 = pd.DataFrame(mat)
synth_df3.columns = df.columns
```
